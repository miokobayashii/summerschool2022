### Obtain the local minimum values (at A and B) of a quartic function by gradient descent $$y = x^4 -1.8 x^3 - 2x^2 + 2x + 2$$
<img src="https://docs.google.com/uc?id=17WzeC9n6khUzJoirXZO0TKI_gyTQzVxj" width="500">


```python
def f(x):
  return x*x*x*x -1.8*x*x*x - 2*x*x + 2*x + 2

def Df(x):
  return 4*x*x*x - 1.8*3 * x*x - 4*x + 2
```

```python
x_series = []
y_series = []

def Exe():

  # initial value
  x = 2.3
  learning_rate = 0.0001

  for i in range(50):

    y_series.append(f(x))
    x_series.append(x)
    x = x - learning_rate * Df(x)
  
```

```python
Exe()
```

```python
x_series, y_series 
```

```python
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import numpy as np


xset = np.arange(-1.5, 2.5, 0.01)
yset = f(xset)
fig, ax = plt.subplots(facecolor="w", figsize = (10,10))
ax.plot(xset, yset, label="f(x)")
ax.plot(x_series, y_series, 'r-o', label="GD")
plt.legend()
```
