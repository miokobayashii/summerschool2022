```python
import numpy as np
```

```python
y = np.array([1,1,0,0])
y_h = np.array([0.99,0.92,0.03,0.1])

y*np.log(y_h) + (1-y)*np.log(1-y_h)
```

```python
E_w = - (np.sum(y*np.log(y_h) + (1-y)*np.log(1-y_h)))/y.shape[0]
print(E_w)
```
