```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np


from keras.layers import Dense, Flatten, Reshape
from keras.layers.advanced_activations import LeakyReLU
from keras.models import Sequential
#from keras.optimizers import Adam
from keras.optimizers import adam_v2

from tensorflow.keras import initializers
from tensorflow.keras.utils import plot_model
```

We use a 2x2 pixel image


```python
img_rows = 2
img_cols = 1
channels = 1

img_shape = (img_rows, img_cols, channels)    
```

```python
input_mat = np.array([
    [ [-1] , 
      [0.5] ] 
])
```

```python
model = Sequential()

# Flatten the input image
model.add(Flatten(input_shape=img_shape))

# Fully connected layer
#w_value = [-1, -1, -1, -1, -1, -1, -1, -1]
w_value = [0, 1, 2, 3, -5, -10, -10, -10]
b1_value = [1,1,1,1]

w_initializer= initializers.Constant(value=w_value)
b1_initializer= initializers.Constant(value=b1_value)

model.add(Dense(4, kernel_initializer=w_initializer, bias_initializer=b1_initializer))

# Leaky ReLU activation
model.add(LeakyReLU(alpha=0.01))

# Output layer with sigmoid activation
#v_value = [1, 1, 1, 1]
#b2_value = [0]
#v_initializer= initializers.Constant(value=v_value)
#b2_initializer= initializers.Constant(value=b2_value)
#model.add(Dense(1, activation='sigmoid', kernel_initializer=v_initializer, bias_initializer=b2_initializer))

model.build()
```

```python
model.summary()
plot_model(model, to_file='model.png', show_shapes=True)
```

```python
out = model.predict(input_mat)
```

```python
out.shape
print(out.reshape((out.shape[0],out.shape[1])))
```

```python
out = out.reshape((out.shape[0],out.shape[1]))
np.sum(out)
```

```python
def sigmoid(x):
  return 1.0 / (1.0 + np.exp(-x))
```

```python
sigmoid(np.sum(out))
```
